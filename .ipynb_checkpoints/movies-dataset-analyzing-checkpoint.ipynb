# Genre:

<h4 style="color: #4d0026; font-weight: 600; font-family: sans-serif;">
            Based on the above analysis, we have observed that the "English"
            language is predominantly used in the film industry. Now, we are
            interested in learning more about the movie genres that are

            trends in the film industry.

```python
data.columns
```

```python
data = data[data['orig_lang'].str.contains("English")]
data.head(2)
```

```python
gen_data = data
```

```python
gen_data.info()
```

```python
gen_data.head(2)
```

```python
data['genre'].value_counts()
```

```python
gen_data['genre'][1]
```

```python
gen_data['genre'] = gen_data['genre'].str.replace('\xa0', '')
gen_data['genre'][0]
```

```python
i = 0
strlist = []  # Use list to store genres, including duplicates
for item in gen_data['genre']:
    genres_list = item.split(',')  # Split the 'genre' string into a list of genres
    for gen in genres_list:
        strlist.append(gen.strip())  # Append the genre to the list, strip() removes any leading/trailing whitespaces
        i += 1
```

```python
strlist
```

```python
from collections import Counter
genre_counts = Counter(strlist)
genre_counts
```

```python
type(genre_counts)
```

```python
genre_counts.keys()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Get the genre names and counts from the genre_counts Counter
genres = list(genre_counts.keys())
counts = list(genre_counts.values())

# Calculate the total count of all genres
total_count = sum(counts)

# Set the figure size (width=20 inches, height=10 inches)
plt.figure(figsize=(20, 10))

# Create a bar plot using seaborn
sns.barplot(x=genres, y=counts)

# Add percentage values on top of the bars
ax = plt.gca()
for i, count in enumerate(counts):
    percentage = (count / total_count) * 100
    ax.text(i, count, f'{percentage:.1f}%', ha='center', va='bottom')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from collections import Counter


# Get the genre names and counts from the genre_counts Counter
genres = list(genre_counts.keys())
counts = genre_counts.most_common(5)

# Extract the top 5 genres and their counts
top_5_genres, top_5_counts = zip(*counts)

# Calculate the total count of all genres
total_count = sum(top_5_counts)

# Create a pie chart
plt.figure(figsize=(8, 8))  # Optional: Set the figure size

# Calculate percentage for each genre
percentages = [(count / total_count) * 100 for count in top_5_counts]

# Create the pie chart with percentage labels
plt.pie(top_5_counts, labels=top_5_genres, autopct='%1.1f%%', startangle=140)

# Set the title of the plot
plt.title('Top 5 Genre in English Movies Percenrage')

# Show the plot
plt.show()
```

<h4 style="color: #60371b; font-weight: 600; font-family: sans-serif;">
            Based on the above analysis, we have discovered that there are five movie genres that are highly popular and in demand among the public. These genres have garnered significant attention and interest from audiences, making them the most sought-after categories in the film industry. Filmmakers and production houses can use this information to focus on creating movies within these popular genres to meet public demands and maximize their chances of success.
        </h4>

## Score and Genre

```python
sr_data = data
sr_data.head(2)
```

```python
genre_counts
```

```python
import pandas as pd


top_genre = ['Action','Drama','Comedy','Thriller','Adventure']

gen_score = pd.DataFrame()

for genre in top_genre:
    genre_scores = sr_data[sr_data['genre'].str.contains(genre)]
    genre_mean_score = genre_scores['score'].mean()
    gen_score = pd.concat([gen_score, pd.DataFrame({'genre': [genre], 'score': [genre_mean_score]})], ignore_index=True)

print(gen_score)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Calculate the total count of all scores
total_count = gen_score['score'].sum()

# Set custom colors for the bar plot
custom_colors = ['#cc0066','#1a000d', '#4d0026',  '#99004d', '#660033']

# Create the bar plot
sns.barplot(x=gen_score['genre'], y=gen_score['score'], palette=custom_colors)

# Set labels and title
plt.xlabel("Genre", fontsize=20, color='blue')
plt.ylabel("Rating", fontsize=20, color='green')
plt.title('Genre wise ratings', fontsize=24, color='#4d0066')

# Add text labels for percentage above each bar
ax = plt.gca()
for i, score in enumerate(gen_score['score']):
    percentage = (score / 10)
    ax.text(i, score, f'{percentage:.1f}', ha='center', va='bottom', fontsize=14)

# Show the plot
plt.show()
```

```python
# "score" , 'Budget' and 'revenue' Relation 
br_data = data
br_data.head(2)
```

```python
# Find unique years for all English movies
unique_years = br_data['date_x'].str[-5:].unique()  # Use [-5:] to get the last 4 characters (the year).

movies = pd.DataFrame()

for year in unique_years:
    total_budget = 0
    total_revenue = 0
    eng_movie = br_data[br_data['date_x'].str.contains(year, case=False)]
    
    for index, movie in eng_movie.iterrows():
        movie['budget_x'] = pd.to_numeric(movie['budget_x'])
        movie['revenue'] = pd.to_numeric(movie['revenue'])
        total_budget += movie['budget_x']
        total_revenue += movie['revenue']
        
    # Append data to 'movies' DataFrame for each year
    new_row = {'Year': year, 'Total Budget': total_budget, 'Total Revenue': total_revenue}
    movies = pd.concat([movies, pd.DataFrame([new_row])], ignore_index=True)


unique_years = br_data['date_x'].str[-5:].unique()

movies = movies[movies['Year'] >='2000']
movies.sort_values(by='Year', ascending=True, inplace=True)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.ticker as ticker


plt.figure(figsize=(10, 5))

plt.plot(movies['Year'], movies['Total Budget'], label='Total Budget', linewidth=3)
plt.plot(movies['Year'], movies['Total Revenue'], label='Total Revenue',linewidth=3)

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Total Budget vs Total Revenue')

# Adding a legend to differentiate the two lines
plt.legend()


def crore_formatter(x, pos):
    return f'{x/1e7:.0f} Cr'


# Adding ticker to show the year on the x-axis as integers
plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(20))
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(crore_formatter))

# Manually reverse the tick labels to display in descending order
# plt.gca().invert_xaxis()

# Show the plot
plt.show()
```

<li style="color: black;">
                Based on the above graph, we can observe the year-wise trends in
            movie budgets and revenues from 2000 to 2023. It shows that over
            this period, both movie budgets and revenues have fluctuated, with
            some years experiencing an increase, while others have seen a
            decrease.
            Notably, from 2020 to 2023, the movie industry has seen an overall
            upward trend in profits and revenues. Even during the years with
            budget fluctuations, the movies have managed to generate profits,
            indicating that the film industry has remained profitable during
            this time span.
            This data suggests that the movie industry has shown resilience and
            profitability over the years, with a positive trend in revenue
            generation. Filmmakers and investors can use this information to
            gain insights into the overall performance of the film industry and
            make informed decisions about their movie projects.
        </li>

```python
srb = data
srb.head(2)

```

```python
unique_score = srb['score'].value_counts()
```

```python
import pandas as pd

srb_x = pd.DataFrame()
unique_score = srb['score'].value_counts()
score_values = unique_score.index

for item in score_values:
    total_budget = 0
    total_revenue = 0
    new_srb = srb[srb['score'] == item]
    
    for index, srb_data in new_srb.iterrows():  # Iterate over the new_srb DataFrame
        srb_data['budget_x'] = pd.to_numeric(srb_data['budget_x'])
        srb_data['revenue'] = pd.to_numeric(srb_data['revenue'])
        total_budget += srb_data['budget_x']
        total_revenue += srb_data['revenue']
      
    new_entry = {'score': item, 'Total Budget': total_budget, 'Total Revenue': total_revenue}
    srb_x = pd.concat([srb_x, pd.DataFrame([new_entry])], ignore_index=True)

```

```python
srb_x.sort_values(by='score', ascending=True, inplace=True)

srb_x.head(2)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.ticker as ticker

plt.figure(figsize=(10, 5))

plt.plot(srb_x['score'], srb_x['Total Budget'], label='Total Budget', linewidth=2)
plt.plot(srb_x['score'], srb_x['Total Revenue'], label='Total Revenue',linewidth=2)

# Adding labels and title
plt.xlabel('Score',color='blue')
plt.ylabel('Amount',color='blue')
plt.title('Total Budget vs Total Revenue',color='#4d0026',fontsize=19)

# Adding a legend to differentiate the two lines
plt.legend()


def crore_formatter(x, pos):
    return f'{x/1e7:.0f} Cr'


# Adding ticker to show the year on the x-axis as integers
plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(5))
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(crore_formatter))

# Show the plot
plt.show()
```
