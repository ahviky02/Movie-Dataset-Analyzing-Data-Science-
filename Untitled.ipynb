```python
import wikipediaapi

def fetch_movie_info(movie_title):
    wiki_wiki = wikipediaapi.Wikipedia(
        language='en',
        extract_format=wikipediaapi.ExtractFormat.WIKI,
        user_agent="Your User Agent"
    )
    page = wiki_wiki.page(movie_title)

    if page.exists():
        return page.summary
    else:
        return f"Movie '{movie_title}' not found on Wikipedia."

# Replace with the movie title you want to fetch information for
movie_title = "Mummies"

movie_info = fetch_movie_info(movie_title)
print(movie_info)
```

```python
import wikipediaapi

def fetch_movie_cast(movie_title):
    wiki_wiki = wikipediaapi.Wikipedia(
        language='en',
        extract_format=wikipediaapi.ExtractFormat.WIKI,
        user_agent="Your User Agent"
    )
    page = wiki_wiki.page(movie_title)

    if page.exists():
        # Extract the section containing cast information
        cast_section = page.section_by_title("Cast")

        if cast_section:
            # Extract the text content of the cast section
            cast_text = cast_section.text
            return cast_text
        else:
            return f"Cast information not found for '{movie_title}' on Wikipedia."
    else:
        return f"Movie '{movie_title}' not found on Wikipedia."

# Replace with the movie title you want to fetch cast information for
movie_title = "Avengers: Endgame"

cast_info = fetch_movie_cast(movie_title)
print(cast_info)
```

```python
import wikipediaapi

def fetch_movie_starring(movie_title):
    wiki_wiki = wikipediaapi.Wikipedia(
        language='en',
        extract_format=wikipediaapi.ExtractFormat.WIKI,
        user_agent="Your User Agent"
    )
    page = wiki_wiki.page(movie_title)

    if page.exists():
        # Search for cast information keywords within the page sections
        cast_keywords = ["cast", "starring", "actors"]
        for section in page.sections:
            section_text = section.text.lower()
            for keyword in cast_keywords:
                if keyword in section_text:
                    return section.text
        
        return f"Starring information not found for '{movie_title}' on Wikipedia."
    else:
        return f"Movie '{movie_title}' not found on Wikipedia."

# Replace with the movie title you want to fetch starring information for
movie_title = "Avengers: Endgame"

starring_info = fetch_movie_starring(movie_title)
print(starring_info)
```

```python
import requests

# Replace with your TMDb API key
api_key = "YOUR_TMDB_API_KEY"

# Replace with the movie title you want to fetch the poster for
movie_title = "Avengers: Endgame"

# Search for the movie using TMDb API
search_url = f"https://api.themoviedb.org/3/search/movie"
params = {
    "api_key": api_key,
    "query": movie_title
}
response = requests.get(search_url, params=params)
data = response.json()

# Get the movie details including the poster path
if "results" in data and data["results"]:
    movie_id = data["results"][0]["id"]
    poster_path = data["results"][0]["poster_path"]
    poster_url = f"https://image.tmdb.org/t/p/original{poster_path}"
    print("Movie Poster URL:", poster_url)
else:
    print("Movie not found.")
```
